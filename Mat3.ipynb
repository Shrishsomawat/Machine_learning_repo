```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
days=[1,2,3,4,5,6,7]
max_t=[56,65,55,53,54,52,50]
min_t=[23,32,19,24,39,33,30]
avg_t=[45,48,40,43,46,44,42]
```

```python
plt.xlabel('days')
plt.ylabel('temprature')
plt.title('chart of weather')

plt.plot(days,max_t,label="Max")
plt.plot(days,min_t,label="Min")
plt.plot(days,avg_t,label="avg")

plt.legend(loc="best",shadow=True)
plt.grid()
```
